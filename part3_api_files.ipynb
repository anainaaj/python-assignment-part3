# Task 1 — File Read & Write Basics 

```python
#To define our notes as strings
notes = [
    "Topic 1: Variables store data. Python is dynamically typed.",
    "Topic 2: Lists are ordered and mutable.",
    "Topic 3: Dictionaries store key-value pairs.",
    "Topic 4: Loops automate repetitive tasks.",
    "Topic 5: Exception handling prevents crashes.",
]

#Writing the above five lines to the file using write mode
with open("python_notes.txt", "w", encoding="utf-8") as f:
    for line in notes:          # Loop through each item in list
        f.write(line + "\n")      # Write the line to the file, adding a newline character
print("File written successfully.")

#Append 2 more lines using append mode
extra_notes = [
    "Topic 6: Functions help us reuse code without repetition.",
    "Topic 7: Modules let us import pre-built tools into Python.",
]

with open("python_notes.txt", "a", encoding="utf-8") as f:
    for line in extra_notes:
        f.write(line + "\n")

print("Lines appended successfully.")
```

```python
## PART B - READ
#Reading the file back and printing each line numbered

with open("python_notes.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()  # Read all lines into a list

for number, line in enumerate(lines, start=1):  # Enumerate lines starting from 1
    print(f"{number}. {line.strip()}")  # Print the line number and the line content without extra whitespace

#To count and print total lines
print(f"Total lines: {len(lines)} ")    
```

```python
## KEYWORD SEARCH
keyword = input(" Enter a keyword to search for: ") 

matches = []  # An empty list to collect matching lines

for line in lines:
    if keyword.lower() in line.lower():   #check if the keyword is in the line (case-insensitive)
        matches.append(line.strip())      # If it matches, add the line to the matches list

#To check results
if matches: 
    print(f"\n Found {len(matches)} matching line(s) for '{keyword}':")
    for match in matches:
        print(f"  → {match}")
else:        # Runs if no matches were found (list is empty)
    print(f"\n No lines found containing '{keyword}'. Try a different keyword!")
```

## Task 2 — API Integration

```python
import requests
import json
from datetime import datetime
import os

BASE_URL = "https://dummyjson.com/products"
print("Libraries imported successfully!")
print(f"🌐 Base API URL: {BASE_URL}")
```

```python
#Fetch and Display 20 products
url = f"{BASE_URL}?limit=20" # Constructing the URL with a query parameter to limit results to 20
response = requests.get(url, timeout=5)  # GET request to the API

data = response.json()  # Parse the JSON response into a Python dictionary
products = data["products"] 

print(f" Fetched {len(products)} products from API\n")

#To print a formatted table
print(f"{'ID':<4} | {'Title':<35} | {'Category':<15} | {'Price':>8} | {'Rating':>6}")
print("-" * 80)

# Loop through each product in list
for product in products:
    pid      = product["id"]           
    title    = product["title"][:33]  
    category = product["category"][:13] 
    price    = product["price"]        
    rating   = product["rating"]       
print(f"{pid:<4} | {title:<35} | {category:<15} | ${price:>7.2f} | {rating:>6.2f}")
```

```python
# Step 2: Filter and Sort
high_rated = [p for p in products if p["rating"] >= 4.5] # List comprehension to filter products with rating >= 4.5
sorted_products = sorted(high_rated, key=lambda p: p["price"], reverse=True) # Sorting the filtered products by price in descending order

print(f"{len(high_rated)} products have rating ≥ 4.5 (sorted by price, highest first):\n")
print(f"{'ID':<4} | {'Title':<35} | {'Price':>8} | {'Rating':>6}")
print("-" * 65)

for p in sorted_products:
    print(f"{p['id']:<4} | {p['title'][:33]:<35} | ${p['price']:>7.2f} | {p['rating']:>6.2f}")
```

```python
# Step 3: Search by category (laptops)
category_url = f"{BASE_URL}/category/laptops"

response_cat = requests.get(category_url, timeout=5)
laptops_data = response_cat.json()
laptops = laptops_data["products"]

print(f"LAPTOPS found: {len(laptops)}\n")
print(f"{'#':<3} | {'Name':<40} | {'Price':>8}")
print("-" * 60)

#  Loop through laptops and print their name and price in a formatted way
for i, laptop in enumerate(laptops, start=1):
    print(f"{i:<3} | {laptop['title'][:38]:<40} | ${laptop['price']:>7.2f}")
```

```python
## Step 4: POST REQUEST (Create a product)

#add new product data in python dictionary format
new_product = {
    "title": "My Custom Product",
    "price": 999,
    "category": "electronics",
    "description": "A product I created via API"
}
post_response = requests.post(f"{BASE_URL}/add", json=new_product, timeout=5)

# Parse the JSON response
result = post_response.json()

print("POST Request sent!")
print(f"Server Response (status code: {post_response.status_code}):")
print("-" * 50)

print(json.dumps(result, indent=2)) 

```

## Task 3 — Exception Handling 

```python
## GUARDED CALCULATOR
def safe_divide(a, b):
    """
    Divides a by b safely.
    Returns result OR an error message string.
    
    The triple-quoted text above is called a DOCSTRING —
    it documents what the function does.
    """
    try:
    
        result = a / b
        return result  

    except ZeroDivisionError:
        # Triggered when b == 0 
        return "Error: Cannot divide by zero"

    except TypeError:
        # Triggered when a or b is the wrong TYPE
        return "Error: Invalid input types"
    
test_cases = [
    (10, 2),        # Normal division 
    (10, 0),        # Division by zero → ZeroDivisionError
    ("ten", 2),    # Wrong type → TypeError
]

print(" SAFE DIVIDE TEST RESULTS")
print("-" * 40)

for a, b in test_cases:  # Unpacking
    result = safe_divide(a, b)
    print(f"safe_divide({a!r}, {b!r}) → {result}")
```

```python
## GUARDED FILE READER

def read_file_safe(filename):
    """
    Safely reads a file and returns its content.
    Handles missing files gracefully.
    """
    try:
        #To Try to open and read the file
        with open(filename, "r", encoding="utf-8") as f:
            content = f.read()   
        return content
    except FileNotFoundError:
        # Triggered when the file doesn't exist at the specified path
        print(f"Error: File '{filename}' not found.")
        return None
    finally:
         print(f" File operation attempt complete for: '{filename}'")

 ## TEST 1 : File that EXISTS
print("TEST 1: Reading 'python_notes.txt' (exists)")
print("=" * 55)
content = read_file_safe("python_notes.txt")
if content:  
    print(f"\n File content (first 100 chars): {content[:100]}...")

print()

# TEST 2: File that DOES NOT EXIST 

print("TEST 2: Reading 'ghost_file.txt' (doesn't exist)")
print("=" * 55)
content2 = read_file_safe("ghost_file.txt")
if content2 is None:
    print("(No content returned — file was missing)")
```

```python
## Part C : ROBUST API CALLS
def fetch_products_safe(limit=20):
    """Fetches products with full error handling."""
    url = f"{BASE_URL}?limit={limit}"
    try:
        response = requests.get(url, timeout=5) 
        #raising an exception for 4xx/5xx HTTP errors
        response.raise_for_status()
        return response.json()["products"]

    except requests.exceptions.ConnectionError:
        print(" Connection failed. Please check your internet.")
        return []

    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
        return []

    except Exception as e:
        # Catches any other unexpected error
    
        print(f"Unexpected error: {e}")
        return []


def post_product_safe(product_data):
    """Posts a new product with full error handling."""
    try:
        response = requests.post(f"{BASE_URL}/add", json=product_data, timeout=5)
        response.raise_for_status()
        return response.json()

    except requests.exceptions.ConnectionError:
        print(" Connection failed. Please check your internet.")
        return None

    except requests.exceptions.Timeout:
        print(" Request timed out. Try again later.")
        return None

    except Exception as e:
        print(f" Unexpected error: {e}")
        return None


#TEST THE SAFE FUNCTIONS -
print("Testing robust fetch_products_safe()...")
safe_products = fetch_products_safe(limit=5)
if safe_products:
    print(f" Got {len(safe_products)} products safely")
    for p in safe_products:
        print(f"  • {p['title']} — ${p['price']}")

print()
print("Testing robust post_product_safe()...")
test_product = {"title": "Robust Test Product", "price": 42, "category": "test"}
post_result = post_product_safe(test_product)
if post_result:
    print(f" Product created with ID: {post_result.get('id', 'N/A')}")
```

```python
## Part D: INPUT VALIDATION LOOP
print("🔎 PRODUCT LOOKUP (Enter ID 1–100, or 'quit' to exit)")

while True:
    
    user_input = input("\nEnter a product ID (1–100), or 'quit' to exit: ").strip()
 
 #Checking for quit first
    if user_input.lower() == "quit":
        print("Exiting product lookup. Goodbye!")
        break

    
    try:
          product_id = int(user_input)
    except ValueError:
        print(f"'{user_input}' is not a valid number. Please enter a number between 1 and 100.")
        continue  
    if not (1 <= product_id <= 100):
        print("Please enter a number between 1 and 100.")
        continue

 # making an API call — input is validated
    try:
        lookup_url = f"{BASE_URL}/{product_id}"
        resp = requests.get(lookup_url, timeout=5)

        #CHECKING STATUS CODE 
        if resp.status_code == 404:
            print(f" Product ID {product_id} not found (404).")
        elif resp.status_code == 200:
            product = resp.json()
            print(f"Found: {product['title']} — ${product['price']}")
        else:
            print(f"Unexpected status code: {resp.status_code}")

    except requests.exceptions.ConnectionError:
        print(" Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print(f" Error: {e}")   
```

## Task 4 — Logging to File 

```python
LOG_FILE = "error_log.txt"

def log_error(function_name, error_type, error_message):
    """
    Writes a timestamped error entry to the log file.
    
    Parameters:
        function_name  (str): Which function had the error
        error_type     (str): Type of error (e.g., 'ConnectionError')
        error_message  (str): What went wrong
    """
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    #Building the log entry in required format
    log_entry = f"{timestamp} | {function_name} | {error_type} | {error_message}\n"
    
    with open(LOG_FILE, "a", encoding="utf-8") as log_file:
        log_file.write(log_entry) 
    
    #Also printting the log entry to console for immediate feedback
    print(f"Logged: {log_entry.strip()}")


print("log_error() function defined and ready!")
```

```python
def trigger_errors():
   
    #Triggering a connection error
    print("Attempting to reach a non-existent URL...")
    fake_url = "https://this-host-does-not-exist-xyz.com/api" 
    try: 
        response = requests.get(fake_url, timeout = 2)
    except requests.exceptions.ConnectionError as e:
        log_error("trigger_errors", "ConnectionError", "No connection could be made")
        print("Logged: ConnectionError")
    except Exception as e: 
        log_error("trigger_errors", "UnexpectedError", str(e))
   
    #Triggering a HTTP 404 Error(Manual Check)
    print("Requesting a non-existent product ID(99)...")
    product_url = "https://dummyjson.com/products/999"
    try:
        response = requests.get(product_url, timeout=5)
        if response.status_code != 200:
            log_error("trigger_errors", "HTTPError", f"{response.status_code} Not Found for product ID 999")
            print(f"Logged: HTTP{response.status_code} Error")
    except Exception as e:
        log_error("trigger_errors", "RequestError", str(e))

def display_logs():
    print("\n--- Current Contents of error_log.txt ---")
    if os.path.exists(LOG_FILE):
        with open(LOG_FILE, "r", encoding="utf-8") as f:
            content = f.read()
            print(content if content else "Log file is empty.")
    else: 
        print("Log file not found.")
if __name__ == "__main__":
    trigger_errors()
    display_logs()
```
